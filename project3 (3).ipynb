```python
#טעינה והעלאת הקובץ ממנו אנתח מידע כמו גם הצגת 3 הרשומות הראשונות
```

```python
df = pd.read_csv('FIFA World Cup 1930-2022 All Match Dataset.csv', encoding="latin1")
df.head(3)
```

```python
#הצגת מידע כללי אודות הטבלה שהועלתה
```

```python
df.info()
```

```python
df.describe()
```

```python
#הצגת העמודות בהם ארצה לשנות את הערכים החסרים למלל(no)
```

```python
collums_to_change = ['Group Stage', 'Knockout Stage', 'Replayed', 'Replay', 'Extra Time',  'Penalty Shootout',
    'Home Team Win',
    'Away Team Win',
    'Draw'
]
```

```python
df[collums_to_change] = df[collums_to_change].replace(0, np.nan)
df[collums_to_change] = df[collums_to_change].fillna('no')
df.head(2)
```

#המרת המקף בתאריכי המשחקים לקו נטוי

```python
df['Match Date'].head(2)
```

```python
df['Match Date'] = pd.to_datetime(df['Match Date']).dt.strftime('%Y/%m/%d')
df['Match Date'].head(2)
```

```python
df[['Home Team Score','Away Team Score']].head(2)
```

```python
#יצירת  עמודה חדשה(סה"כ שערים פר משחק) על ידי חיבור של שתי עמודות
```

```python
df['TotalGoals'] = df['Home Team Score'] + df['Away Team Score']
```

```python
df[['TotalGoals']].head(2)
```

```python
#המשחק בו הובקעו הכי הרבה שערים בהיסטוריה של המונדיאלים
```

```python
Max_goals_in_a_game = df['TotalGoals'].max()
Max_goals_in_a_game
```

```python
#ממוצע השערים למשחק במונדיאל מאז החלו הטורנירים
```

```python
Avg_goals_in_a_game = df['TotalGoals'].mean()
Avg_goals_in_a_game
```

```python
# במונדיאלים לאורך הסטיית התקן של כמות השערים שמובקעת פר משחק
```

```python
df['TotalGoals'].std()
```

```python
#המודיאל בו כמות השערים הממוצעת הכוללת היתה הנמוכה ביותר  
```

```python
Avg_goals_in_a_game = df.groupby('tournament Name')['TotalGoals'].agg(['mean'])

Avg_goals_in_a_game = Avg_goals_in_a_game.sort_values('mean', ascending=True)

Avg_goals_in_a_game.head(1)
```

```python
#המודיאל בו כמות השערים הממוצעת הכוללת היתה הגבוהה ביותר  
```

```python
Avg_goals_in_a_game.tail(1)
```

```python
#כמות השערים שהובקעה בכל מונדיאל (מהמונדיאל הראשון ועד לאחרון)
```

```python
total_goals_by_tournament = df.groupby('tournament Name')['TotalGoals'].sum()

total_goals_by_tournament.tail(10)
```

```python
#ניתוח של מספר המשחקים הכולל ששוחקו בהיסטוריה של המונדיאלים,כמות שערים פר משחק ,וסך השערים שהובקעו בכל המונדיאלים
```

```python
df.agg(
    TotalWorldCupGoals=('TotalGoals', 'sum'),
    TotalWorldCupMatches=('TotalGoals', 'count'),
    AvgGoalsPerMatch=('TotalGoals', 'mean')
)
```

```python
#מספר השערים הגבוה ביותר שכבשה קבוצה במשחק(בית ו"חוץ")
```

```python
df.agg(
    MaxHomeScore=('Home Team Score', 'max'),
    MaxAwayScore=('Away Team Score', 'max')
).fillna(0)
```

### גרפים

```python
#גרף 1 קו – ממוצע שערים למשחק בכל מונדיאל
```

```python
avg_goals_tour = df.groupby("tournament Name")["TotalGoals"].mean()

plt.figure(figsize=(12, 5))#רזולוציית הגרף
plt.plot(avg_goals_tour.index, avg_goals_tour.values)
plt.title("Average Goals per Match by Tournament")#כותרת לגרף
plt.xlabel("Tournament Name")#כותרת ציר X
plt.ylabel("Avg Goals per Match")#כותרת ציר Y
plt.xticks(rotation=90)#עיצוב וסידור המלל שבציר האיקס
plt.tight_layout()#סידור וריווח נוסף של העמודות והמלל שבגרף
plt.grid(True)#הוספת עיצוב משובץ לגרף

plt.show()#הצגת הגרף 
```

```python
#גרף 2- עמודות – סך כל השערים בכל מונדיאל
```

```python
total_goals_tour = df.groupby("tournament Name")["TotalGoals"].sum()

plt.figure(figsize=(12, 5))
plt.bar(total_goals_tour.index, total_goals_tour.values)
plt.title("Total Goals per Tournament")
plt.xlabel("Tournament Name")
plt.ylabel("Total Goals")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
 # הגרף 3- היסטוגרמה – התפלגות מספר השערים במשחק
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df["TotalGoals"], bins=range(0, df["TotalGoals"].max() + 2))
plt.title("Distribution of Goals per Match")
plt.xlabel("Goals per Match")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
#גרף 4-גרף עמודות אופקי(5 המשחקים עם הכי הרבה שערים בהיסטוריה)
```

```python
top_matches = df.nlargest(5, "TotalGoals")

plt.figure(figsize=(10, 6))
plt.barh(top_matches["Match Name"], top_matches["TotalGoals"])
plt.title("Top 5 Highest Scoring Matches")
plt.xlabel("Total Goals")
plt.tight_layout()
plt.show()
```

```python
#גרף 5 -נקודות(מתאר את הקשר שבין כמות המשחקים לטורניר לכמות השערים שהובקעו בטורניר)
```

```python
games = df.groupby("tournament Name").size()
goals = df.groupby("tournament Name")["TotalGoals"].sum()

plt.figure(figsize=(8,6))
plt.scatter(games, goals)
plt.title("Games vs Total Goals per Tournament")
plt.xlabel("Number of Games")
plt.ylabel("Total Goals")
plt.grid(True)
plt.show()
```
